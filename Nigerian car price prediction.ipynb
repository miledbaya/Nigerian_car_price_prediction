question5:Handle missing values, if any, by analyzing and deciding on an appropriate strategy.

```python
data.isnull().sum()
```

```python
data['Year of manufacture'].value_counts()
```

```python
data['Year of manufacture']=data['Year of manufacture'].fillna(2007)
```

```python
data.isnull().sum()
```

```python
data['Condition'].value_counts()
```

```python
data['Condition']=data['Condition'].fillna('Nigerian Used')
```

```python
data.isnull().sum()
```

```python
data['Mileage'].fillna(data['Mileage'].mean(),inplace=True)
data['Engine Size'].fillna(data['Engine Size'].mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Fuel'].value_counts()
```

```python
data['Fuel']=data['Fuel'].fillna('Petrol')
```

```python
data.isnull().sum()
```

```python
data['Transmission'].value_counts()
```

```python
data['Transmission']=data['Transmission'].fillna('Automatic')
```

```python
data.isnull().sum()
```

```python
data['Build'].value_counts()
```

```python
data['Build']=data['Build'].fillna('SUV')
```

question6:

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

file_path = 'Nigerian_Car_Prices.csv'
data = pd.read_csv(file_path)


plt.figure(figsize=(12, 4))


plt.subplot(1, 3, 1)
sns.histplot(data=data, x='Year of manufacture', bins=20, kde=True)
plt.title('Year of Manufacture')


plt.subplot(1, 3, 2)
sns.histplot(data=data, x='Mileage', bins=20, kde=True)
plt.title('Mileage')


plt.subplot(1, 3, 3)
sns.histplot(data=data, x='Engine Size', bins=20, kde=True)
plt.title('Engine Size')


plt.tight_layout()
plt.show()


```

question7:

```python
mileage_percentile = data['Mileage'].quantile(0.99)


data = data[data['Mileage'] <= mileage_percentile]


print(data.head())
```
